```python
import pandas as pd
import numpy as np

# import covid patient data
covid = pd.read_csv('covid.csv')
covid.describe().T
```

```python
covid.head()
```

```python
# Return if the person is died(1) or alive(0)
def died (row):
    if row['date_died'] == '9999-99-99':
        return 0
    return 1

covid['died'] = covid.apply (lambda row: died(row), axis=1)

# Return how many days the person took between first sympton and hospitalizatioin
import datetime
def days_before_hospital (row):
    symptom = row['date_symptoms'].split('-')
    entry = row['entry_date'].split('-')
    days = datetime.datetime(int(entry[2]), int(entry[1]), int(entry[0])) - datetime.datetime(int(symptom[2]), int(symptom[1]), int(symptom[0]))
    return days.days
    
covid['days_before_hospital'] = covid.apply (lambda row: days_before_hospital(row), axis=1)

# drop the unused columns
covid.drop(columns={'entry_date','date_died','date_symptoms','id'},axis=1,inplace=True)
covid.describe().T
```

```python
# Min-Max Normalization
covid = covid[covid['covid_res']!=3]
norm_covid=(covid-covid.min())/(covid.max()-covid.min())
norm_covid
```

```python
# target is tobacco
y = norm_covid['died']
# rest of them are features
norm_covid.pop('died')
X = norm_covid
```

```python
from sklearn.model_selection import train_test_split
# split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=35)
```

```python
# Apply logistic regression model
from sklearn.linear_model import LogisticRegression
# Library for evaluation
from sklearn.metrics import accuracy_score, classification_report
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

lr_pred = logreg.predict(X_test)
print("Accuracy score:", accuracy_score(y_test, lr_pred))
print("\nClassification Report\n")
print(classification_report(y_test, lr_pred))
```

```python
# Apply xgboost model
import xgboost as xgb
D_train = xgb.DMatrix(X_train, label=y_train)
D_test = xgb.DMatrix(X_test, label=y_test)

param = {
    'eta': 0.3, 
    'max_depth': 3,  
    'objective': 'multi:softprob',  
    'num_class': 2} 

steps = 20

xgbModel = xgb.train(param, D_train, steps)

xgb_pred = xgbModel.predict(D_test)
best_pred = np.asarray([np.argmax(line) for line in xgb_pred])

print("Accuracy score:", accuracy_score(y_test, best_pred))
print("\nClassification Report\n")
print(classification_report(y_test, best_pred))
```

```python
#apply random forest classifier
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=500, max_depth=10)
rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)

print("Accuracy score:", accuracy_score(y_test, rf_pred))
print("\nClassification Report\n")
print(classification_report(y_test, rf_pred))
```
